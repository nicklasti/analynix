```python
import yfinance as yf
tickers = ['AADI', 'ABCL', 'ABCM', 'ABEO', 'ABIO']

import asyncio
import time

revenues = []

async def get_symbols(ticker_str: str):
    revenue = yf.Ticker(ticker_str).info['totalRevenue']
    revenues.append(revenue)
    print(f'Added {ticker_str}')

async def poop_fart():
    ticker_tasks = []
    for ticker in tickers:
        ticker_tasks.append(asyncio.create_task(get_symbols(ticker)))
    await asyncio.wait(ticker_tasks)

task = asyncio.create_task(poop_fart())
done, pending = await asyncio.wait({task})
print(revenues)
```
